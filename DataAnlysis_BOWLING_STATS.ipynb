Let's begin by importing`matplotlib.pyplot` and `seaborn`.

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

 **Bar plot for the first 10 wicket-takers**

````python
plt.figure(figsize=(12,6))
plt.bar(cricket_df['Player'].head(10), cricket_df['Wkts'].head(10))
plt.plot(cricket_df['Player'].head(10), cricket_df['Wkts'].head(10), 'o--r')
plt.title('First 10 Wicket Takers in World Cup')
plt.xlabel('Player')
plt.ylabel('Wickets')
plt.xticks(rotation=60)
plt.show()
````

**Number of Players from Each Country in list**

````python

# Number of Players from Each Country in list
player_count = cricket_df.groupby('Team')['Player'].count()
plt.figure(figsize=(12, 6))
sns.barplot(x=player_count.values, y=player_count.index)
plt.ylabel('Team')
plt.xlabel('Number of Players')
plt.title('Number of Players from Each Country')
plt.xticks(rotation=0)
plt.show()


````

**Which bowlers have the best economy rates in the tournament? Visualize the top 10 most economical bowlers using a bar chart.**

````python
# Bowling Economy Rate Analysis:
# - Which bowlers have the best economy rates in the tournament? Visualize the top 10 most economical bowlers using a bar chart.
plt.figure(figsize=(12, 6))
sorted_df = cricket_df.sort_values('Econ', ascending=True).head(10)
plt.bar(sorted_df['Player'], sorted_df['Econ'])
plt.title('Top 10 Economical Bowlers in the World Cup')
plt.xlabel('Player')
plt.ylabel('Economy Rate')
plt.xticks(rotation=45)
plt.show()







````

**Comparison of average wickets taken by players from each country**

````python
# Comparison of average wickets taken by players from each country
avg_wickets_by_country = cricket_df.groupby('Team')['Wkts'].mean()
plt.figure(figsize=(12,6))
sns.barplot(x=avg_wickets_by_country.values, y=avg_wickets_by_country.index)
plt.xlabel('Average Wickets Taken')
plt.ylabel('Team')
plt.title('Average Wickets Taken by Players from Each Country')
plt.xticks(rotation=0)
plt.show()
````

 **How frequently do bowlers take wickets? Visualize the strike rates of the top 10 bowlers with the lowest strike rates using a bar chart.**

````python
# Strike Rate Comparison:
# - How frequently do bowlers take wickets? Visualize the strike rates of the top 10 bowlers with the lowest strike rates using a bar chart.
plt.figure(figsize=(12, 6))
sorted_df = cricket_df.sort_values('SR', ascending=True).head(10)
plt.plot(sorted_df['Player'], sorted_df['SR'])
plt.title('Top 10 Bowlers with Lowest Strike Rates in the World Cup')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.show()
````

# Q3: Is there a correlation between the number of wickets taken and the economy rate of a bowler?

````python
sns.barplot(x=cricket_df['Wkts'],y=cricket_df['Econ'])
plt.xlabel('Wickets')
plt.ylabel('Economy Rate')
plt.title('Correlation between wickets and economy rate')
plt.show()
````

````python


# Calculate the correlation coefficient
correlation = cricket_df['Wkts'].corr(cricket_df['Econ'])

print("Correlation coefficient:", correlation)

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Wkts', y='Econ', data=cricket_df)
plt.title('Correlation between Wickets Taken and Economy Rate')
plt.xlabel('Wickets Taken')
plt.ylabel('Economy Rate')
plt.show()

# Interpretation of the correlation coefficient:
if correlation > 0:
  print("There is a positive correlation: higher wickets taken tend to be associated with higher economy rates.")
elif correlation < 0:
  print("There is a negative correlation: higher wickets taken tend to be associated with lower economy rates.")
else:
  print("There is no significant correlation between wickets taken and economy rate.")
````

Higher number of wickets are taken when there is lower economy rate
